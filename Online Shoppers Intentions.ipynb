## Coming up with a business problem from the data (i.e. what do we want to predict)?

### For now, we want to find the best predictors of the 'Revenue' feature to find out which features are leading to the highest number of buyers.

#### Create a heatmap to show the correlation between all of the variables

```python
response_feats = df.columns.tolist()
```

```python
response_feats[:]
```

```python
# We must make a correlation matrix to be able to visualize feature relationships.
correlation = df[response_feats].corr()
correlation.iloc[:]
```

```python
sns.heatmap(correlation, xticklabels= correlation.columns.values,
            yticklabels = correlation.columns.values, center = 0)
```

### Information gathered from heatmap:

- As shown across the bottom row, the 'PageValues' feature by far has the highest correlation with the 'Revenue' feature. This makes sense, because the pages that have previously visited high-value pages are likely to have the most interest in purchasing.
- 'Administrative', 'ProductRelated', and 'ProductRelated_Duration' are the only other features that appear to have any sort of significant impact on the result of the 'Revenue' feature.
- 'ExitRates' has a slightly significant negative correlation, which makes sense because people who buy something are more likely to visit another page such as related items, related stores, ads, etc.
